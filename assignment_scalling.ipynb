# Techniques to perform Feature Transformation:

Normalization <br>
Standardization <br>
Log Transformation <br>
Robust Scaler <br>
Max Absolute Scaler

# Normalization

```python
min_price=df.price.min()
max_price=df.price.max()
```

```python
min_price
```

```python
max_price
```

```python
df['price_norm_manual']=(df.price-min_price)/(max_price-min_price)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mmx=MinMaxScaler()
```

```python
df['price_norm_sklearn']=mmx.fit_transform(df[['price']])
```

```python
df.head()
```

# standardization

```python
mean_price=df.price.mean()
std=df.price.std()
```

```python
df['price_std_manual']= (df['price']-mean_price)/std
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
df['price_std_sklearn']=scaler.fit_transform(df[['price']])
```

```python
df.head()
```

```python
std_dof=df.price.std(ddof=0)
```

```python
std_dof
```

```python
std_dof1=df.price.std(ddof=1)
```

```python
std_dof1
```

```python
df
```

```python
df['price_std_manual2']= (df['price']-mean_price)/std_dof1
```

```python
df
```

```python
df['price_std_manual2']= (df['price']-mean_price)/std_dof
```

```python
df
```

# Log Transfomation

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
```

```python
fc=FunctionTransformer(np.log1p)
```

```python
df['price_fc']=fc.fit_transform(df[['price']])
```

```python
df
```

```python
df['fc_np']=np.log1p(df[['price']])
```

```python
df
```

# Robust Scaler

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
max=MaxAbsScaler()
```

```python
df['mx_abs']=max.fit_transform(df[['price']])
```

```python
df
```

# Power Transformer (box,cox)

```python
from scipy.stats import boxcox
df['box cox'],_=boxcox(df['price']+1)
```
